```python
import os
import glob
import pandas as pd

# Путь к папке "outputs" на диске C
folder_path = r'C:\outputs'

# Получаем список всех файлов формата excel в папке "outputs"
excel_files = glob.glob(os.path.join(folder_path, '*.xlsx'))

# Проходимся по каждому файлу
for file in excel_files:
    # Открываем файл excel
    df = pd.read_excel(file)
    
    # Создаем пустые списки для каждого столбца
    column_0 = []
    column_2 = []
    column_4 = []
    
    # Проходимся по каждой строке в датафрейме
    for index, row in df.iterrows():
        # Добавляем значение в соответствующий список столбца
        column_0.append(row[0])
        column_2.append(row[2])
        column_4.append(row[4])
    
    # Выводим информацию о минимальных значениях для каждого столбца
    print(f'{file}, {min(column_0)}, {min(column_2)}, {min(column_4)}')

    # Добавляем пробел после вывода информации о каждом файле
    print()
```
